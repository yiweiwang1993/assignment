```python
#Create a plot that shows two columns (features) plotted against each other (e.g. A vs. B) Create a histogram of the dataset using a selected column (feature)
import matplotlib.pyplot as plt
from pylab import *

#print(df.household_expenditure)
#print(df.SP500)
fig=plt.figure()
plt.scatter(df.household_expenditure, df.SP500)
axis = fig.gca() #get current axis
axis.set_title('household_expenditure vs SP500')
axis.set_xlabel('household_expenditure')
axis.set_ylabel('SP500')
fig
```
