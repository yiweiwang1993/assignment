```python
#Make scatterplots of a selected column "featureA" in the list yfeatureA against the featureA in the list xfeatureA. 
#As the names suggest, plot featureA in yfeatureA along the y featureA and the acids in xfeatureA along the x axis. 
#Label the axes with the respective feature name. 
#Set it up as a grid with 3 rows and 2 columns.

xfeatures=['household_expenditure','export_value_index','GDP']
yfeatures=['employment_ratio','population']

#your code here
figure()
print(len(xfeatures))
print(len(yfeatures))
fig1, axes=plt.subplots(figsize=(10,10), nrows=len(xfeatures), ncols=len(yfeatures))
for i, xfeature in enumerate(xfeatures):
    for j, yfeature in enumerate(yfeatures):
        axes[i][j].scatter(df[xfeature],df[yfeature])
        axes[i][j].set_xlabel(xfeature)
        axes[i][j].set_ylabel(yfeature)
fig1.tight_layout()
show(fig1)
```

```python
#The first concept we deal with here is pandas groupby. 
#Group your dataframe dfsub by the values of a particular factor variable.

RandomClass_groupby = df.groupby('RandomClass')
print (type(RandomClass_groupby))
RandomClass_groupby.head()
```

```python
#The groupby function also acts like an object that can be mapped. 
#After the mapping is complete, the rows are put together (reduced) into a larger dataframe. 
#Apply the describe function.

dfrd=RandomClass_groupby.describe()
print (type(dfrd))
dfrd.head(20)
```

```python
#Standard Deviation of each feature

vecs=[]
keys=[]
for key, value in RandomClass_groupby:
    k=key
    v=value.std()
print (k, type(v), v)

#Mean and Standard Deviation indexed by Clump_Thickness
dfbystd=df.groupby('RandomClass').std()
dfbystd.head()

dfbymean=RandomClass_groupby.aggregate(np.mean)
dfbymean.head()
```

```python
featurelist1=['year', 'employment_ratio',
       'household_expenditure', 'export_value_index',
       'energy_use', 'net_portfolio_investment', 'import_value_index',
       'population', 'GDP', 'SP500']

renamedict_std={k:k+"_std" for k in featurelist1}
renamedict_mean={k:k+"_mean" for k in featurelist1}
dfbystd.rename(inplace=True, columns=renamedict_std)
dfbymean.rename(inplace=True, columns=renamedict_mean) 
dfbystd.head()
```
